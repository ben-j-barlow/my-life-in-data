```python
import requests
from bs4 import BeautifulSoup
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt


def scrape_webpage(webpage):
    # Initialize an empty list to store the scraped text
    text_list = []

    # Send a GET request to the webpage
    response = requests.get(webpage)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all text elements within the webpage and extract the text
    text_elements = soup.find_all(text=True)
    webpage_text = ' '.join(text.strip() for text in text_elements if text.strip())

    return webpage_text

        # Append the webpage text to the list
    #    text_list.append(webpage_text)

    #return text_list


def remove_stopwords(text_list, stopwords):
    cleaned_text_list = []

    for text in text_list:
        # Split the text into individual words
        words = text.split()

        # Remove stopwords from the list of words
        cleaned_words = [word for word in words if word.lower() not in stopwords]

        # Join the cleaned words back into a string
        cleaned_text = ' '.join(cleaned_words)

        # Append the cleaned text to the list
        cleaned_text_list.append(cleaned_text)

    return cleaned_text_list


def generate_wordcloud(text, stopwords):
    # Create a WordCloud object with desired configurations
    wordcloud = WordCloud(collocations=True, width=800, height=400, background_color='white', stopwords=set(stopwords))

    # Generate the word cloud from the input text
    wordcloud.generate(text)

    # Display the word cloud using matplotlib
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
import os

def read_files(directory):
    file_dict = {}
    for filename in sorted(os.listdir(directory), key=lambda x: int(x[:-4])):
        if filename.endswith('.txt'):
            filepath = os.path.join(directory, filename)
            with open(filepath, 'r') as file:
                file_contents = file.read()
                file_dict[filename] = file_contents
    return file_dict

# Provide the directory path
directory_path = "/Users/benbarlow/dev/my-life-in-data/education_and_employment/data/module_description"

# Call the function to read files and store in a dictionary
files_dictionary = read_files(directory_path)
```

```python
import string
import re

def remove_extra_spaces(text):
    # Replace multiple spaces, new lines, and tabs with a single space
    cleaned_text = re.sub(r'\s+', ' ', text.replace('\n', ' '))
    return cleaned_text

def remove_punctuation(text):
    # Create a translation table using the string.punctuation string
    translator = str.maketrans('', '', string.punctuation)
    # Apply the translation table to remove punctuation
    text_without_punct = text.translate(translator)
    return text_without_punct
```

```python
text = " ".join(list(files_dictionary.values()))

from nltk.stem import SnowballStemmer
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud

# Tokenize the text into words
tokens = word_tokenize(text)

# Convert stemmed tokens to lowercase
stemmed_tokens = [token.lower() for token in tokens]

# Join the stemmed tokens back into a single string
processed_text = " ".join(stemmed_tokens)

processed_text = re.sub(r'\d+', '', processed_text)
processed_text = remove_extra_spaces(processed_text)
processed_text = remove_punctuation(processed_text)
```
